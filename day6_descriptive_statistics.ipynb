```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

#cleaning
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns=['Cabin'])


df.head()
```

```python
#summary statistics
df.describe()
```

```python
#mean age
df['Age'].mean()
```

```python
#median age
df['Age'].median()
```

```python
#Key metrics
```

```python
#2. standard deviation
#measuring how spread out passenger ages are
df['Age'].std()
```

```python
#3. fare distribution
df['Fare'].describe()
```

Grouped statistics

```python
#1. survival by gender
# calculating survival rate by gender
df.groupby('Sex')['Survived'].mean()
```

female passengers had a high survival rate than male

```python
#survival by passenger class
#calculating survival rate across passenger classes
df.groupby('Pclass')['Survived'].mean()
```

passengers in higher classes were more likely to survive

```python
import pandas as pd
df = pd.read_csv("california_housing_test.csv")
df.head()
```

```python

#population per household
#creating population per household to measure crowding levels
df['population_per_household'] = df['population'] / df['households']
df[['population', 'households', 'population_per_household']].head()
```

```python
#rooms per hosehold
#creating rooms per household to estimate housing space availability
df['rooms_per_household'] = df['total_rooms'] / df['households']
df[['total_rooms', 'households', 'rooms_per_household']].head()
```

Areas with higher rooms per hosehold may indicate less crowded living conditions

```python
#bedrooms ratio
#calculating bedroom ratio to understand housing layout efficiency
df["bedroom_ratio"] = df['total_bedrooms'] / df['total_rooms']
df[['total_bedrooms', 'total_rooms', "bedroom_ratio"]].head()
```

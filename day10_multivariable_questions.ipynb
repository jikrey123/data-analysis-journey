```python
import pandas as pd
import matplotlib.pyplot as plt

#loading data
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

#cleaning data
df = df.drop(columns=['Cabin'])
df['Age'] = df['Age'].fillna(df['Age'])
df['Embarked'] = df['Embarked'].fillna(df['Embarked'])

df.head()
```

TWO VARIABLE QUESTIONS

```python
#survival by gender
df.groupby('Sex')['Survived'].mean()
```

female passengers had a higher survival rate than males

```python
#survival by passenger class
df.groupby('Pclass')['Survived'].mean()
```

first class passengers had higher survival rates than lower classes

THREE VARIABLE THINKING

```python
#survival by passenger gender and class
df.groupby(['Sex', 'Pclass'])['Survived'].mean()
```

across all passenger classes females had higher survival rates than males

VISUAL MULTIVARIATE ANALYSIS

```python
pivot = df.pivot_table(values='Survived', index= 'Pclass', columns='Sex')
pivot.plot(kind='bar')
plt.title('Survival by gender and class')
plt.ylabel("Survival rate")
plt.show()
```

females have a higher survival rate than males across all classes

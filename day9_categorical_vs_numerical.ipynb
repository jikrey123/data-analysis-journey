```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

#cleaning
df = df.drop(columns=['Cabin'])
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] =df['Embarked'].fillna(df['Embarked'].mode()[0])

df.head()
```

CATEGORICAL - SURVIVAL (BAR CHARTS)

```python
#survival by gender
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title('Survival by gender')
plt.ylabel('Survival rate')
plt.show()
```

female passengers have a high survival rate

```python
#surviuval by Passenger class
df.groupby('Pclass')['Survived'].mean().plot(kind ='bar')
plt.title('Survival rate by passenger')
plt.ylabel('Survival rate')
plt.show()
```

Passengers in high classes are more likely to survive

NUMERICAL - SURVIVAL (BOXPLOTS)

```python
#age vs survival
df.boxplot(column='Age', by='Survived')
plt.title('Age distribution by survival')
plt.suptitle("")
plt.show()
```

young passengers are likely to survive

```python
#fare vs survival
df.boxplot(column='Fare', by='Survived')
plt.title('fare distribution by survival')
plt.suptitle("")
plt.show()
```

CATEGORICAL

```python
#survival by embarkation
df.groupby('Embarked')['Survived'].mean().plot(kind='bar')
plt.title('survival rate by Embarkation port')
plt.ylabel('Survival rate')
plt.show()

```

```python
Survival rates varied across embarkation ports
```
